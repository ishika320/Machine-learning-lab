```python
""" Experiment-3:
Compute correlation coefficients between numerical features and interpret the strength and
direction of relationships. Calculate covariance between selected variables and explain how it
differs from correlation."""
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/rashida048/Datasets/master/StudentsPerformance.csv"

df = pd.read_csv(url)

df.head()
```

```python
num_df = df.select_dtypes(include=['int64','float64'])
num_df.head()
```

```python
# Correlation Coefficient
corr_matrix = num_df.corr()
print("Correlation Matrix:\n", corr_matrix)
```

```python
plt.figure(figsize=(6,5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
cov_matrix = num_df.cov()
print("Covariance Matrix:\n", cov_matrix)
```

```python
# Covariance Between Specific Variables
cov_math_reading = num_df['math score'].cov(num_df['reading score'])

print("Covariance (Math vs Reading):", cov_math_reading)
```

```python
""" Experiment-4:
Plot histograms to analyse data distribution and identify skewness or symmetry. Use boxplots
to visualize quartiles, interquartile range, and potential outliers. """
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/rashida048/Datasets/master/StudentsPerformance.csv"

df = pd.read_csv(url)

df.head()
```

```python
num_df = df.select_dtypes(include=['int64','float64'])
num_df.head()
```

```python
# Histogram for All Numerical Features
num_df.hist(figsize=(10,8))
plt.suptitle("Histograms of Numerical Features")
plt.show()
```

```python
# Individual Histogram Example
plt.hist(df['math score'], bins=20)
plt.title("Math Score Distribution")
plt.xlabel("Math Score")
plt.ylabel("Frequency")
plt.show()
```

```python
# Boxplot (Quartiles + Outliers)
# Boxplot for All Numerical Features
plt.figure(figsize=(8,6))
sns.boxplot(data=num_df)
plt.title("Boxplot of Numerical Features")
plt.show()
```

```python
# Individual Boxplot Example
sns.boxplot(y=df['math score'])
plt.title("Math Score Boxplot")
plt.show()
```
